```python
from peppcbench.plot import plot_contact_map, plot_chain_pairwise_analysis
from pathlib import Path
import json
import pickle
```

```python
job_dir = Path("../pepdb/8c5l_C/af3/8c5l_c/")
sample_dir = job_dir / "seed-1_sample-0"
```

```python
extra_output_path = sample_dir / "extra_output.pkl"
extra_output = pickle.load(open(extra_output_path, "rb"))
contact_map = extra_output["contact_probs"]
```

```python
plot_contact_map(contact_map)
```

```python
# We plot samples with the same seed_id

results = []
seed_id = 1

for sample_dir in job_dir.iterdir():
    if not sample_dir.is_dir():
        continue
    if f"seed-{seed_id}_sample-" not in sample_dir.name:
        continue
    confidence_extra_path = sample_dir / "confidences_extra.json"
    confidence_extra = json.load(open(confidence_extra_path, "r"))
    results.append(confidence_extra)
```

```python
plot_chain_pairwise_analysis(results)
```
